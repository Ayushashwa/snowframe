```python
import pandas as pd
from snowframe import SnowFrame, set_active_session

sales = pd.DataFrame({
    "customer_id": [1, 1, 2, 3, 3],
    "amount": [100, 2500, 700, 1800, 300],
})

customers = pd.DataFrame({
    "customer_id": [1, 2, 3],
    "name": ["Amit", "Riya", "Kabir"],
})
```

```python
sales
```

```python
customers
```

```python
sf = SnowFrame()
sf.auto()
set_active_session(sf)

sf.tables()
```

```python
%%sf

select * from customers;
```

```python
%%sf
select * from sales;
```

```python
%%sf
create or replace table final_merged as 
select a.*,b.name from (select * from sales) a
left join (select * from customers) b
on a.customer_id=b.customer_id;
```

```python
%%sf

SELECT *
FROM final_merged;
```
